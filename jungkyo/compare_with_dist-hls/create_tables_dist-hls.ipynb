```python
dist_hls_dir_path = f'../../../dist-s1-research_develop_jung/jungkyo/1_visualize_rtc_time_series/DIST-ALERT_links/'
```

```python
dist_hls_txt_paths = glob.glob(f'{dist_hls_dir_path}/*')
print(dist_hls_txt_paths)
```

```python
HLS_MGRS_id = np.unique([text_file.split('/')[-1].split('.')[0] for text_file in dist_hls_txt_paths])
```

```python
HLS_MGRS_id
```

```python
def extract_info(file_path):
    filename = file_path.split('/')[-1].split('.')[0]
    file_char = filename.split('_')
    mgrs = file_char[3]
    date = file_char[4]
    file_type = file_char[9]
    return date, mgrs, file_type
```

```python
# Example file path

file_path = 'https://glad.umd.edu/projects/opera/SEP/v1sample/DIST-ALERT_v1/2021/10/S/F/J/DIST-ALERT_2021274T185119_S30_T10SFJ_v1/OPERA_L3_DIST-ALERT-HLS_T10SFJ_20211001T185119Z_20231206T155422Z_S2B_30_v1_GEN-DIST-DATE.tif'

# Test the function with the example path
date, mgrs, file_type = extract_info(file_path)
print(f"Date: {date}, MGRS: {mgrs}, Type: {file_type}")
```

```python
data = []

for text_file in dist_hls_txt_paths:
    with open(text_file, 'r') as file:
        for line in file:
            line = line.strip()
            if line.endswith('.tif'):
                if line:
                    date, mgrs, file_type = extract_info(line)
                    if date and mgrs and file_type:
                        data.append([date, mgrs, file_type, line])
```

```python
data
```

```python
df = pd.DataFrame(data, columns=['Date', 'MGRS', 'Type',  'DIST-HLS_path'])
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%dT%H%M%SZ')
```
