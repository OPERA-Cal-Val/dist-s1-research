```python
import geopandas as gpd
from pathlib import Path
from rasterio.crs import CRS
import contextily as ctx
import matplotlib.pyplot as plt
```

```python
def get_centroid(path: Path):
    geo = gpd.read_file(path).geometry.unary_union.centroid
    return geo
```

```python
event_aoi_paths = list(Path('../../../dist_s1_events/external_validation_data_extents_db/').glob('*'))
event_aoi_paths = [p for p in event_aoi_paths if p.stem[0] != '.']
event_aoi_paths
```

```python
geos = list(map(get_centroid, event_aoi_paths))
```

```python
names = [p.stem for p in event_aoi_paths]
def get_type(stem: str) -> str:
    if 'fire' in stem:
        return 'fire'
    if 'landslide' in stem:
        return 'landslide'
    if 'flood' in stem:
        return 'flood'
df_events = gpd.GeoDataFrame({'names': names,
                              'event_type': list(map(get_type, names))
                             },
                             geometry=geos, crs=CRS.from_epsg(4326))
df_events.head()
```

```python
df_events.explore(column='event_type', cmap='tab20c', style_kwds=dict(color='black', weight=1))
```
