```python
# Specify the path to the image file
image_path = "https://opera-pst-rs-pop1.s3.us-west-2.amazonaws.com/products/RTC_S1/OPERA_L2_RTC-S1_T036-076225-IW1_20201006T042403Z_20240426T220152Z_S1B_30_v1.0/OPERA_L2_RTC-S1_T036-076225-IW1_20201006T042403Z_20240426T220152Z_S1B_30_v1.0_VV.tif"  # Replace with the actual path to your image file

# Open the image file
with rasterio.open(image_path) as src:
    # Access information about the image
    print("Image width:", src.width)
    print("Image height:", src.height)
    print("Number of bands:", src.count)
    print("CRS (Coordinate Reference System):", src.crs)
    print("Transform (Affine matrix):", src.transform)
```

```python
# URL of the remote raster file

# Open the remote raster file
with rasterio.open(image_path) as src:
    # Read the raster data
    raster_data = src.read()

    # Visualize the raster using Matplotlib
    plt.imshow(raster_data[0], cmap='gray', vmin=0, vmax=0.5)  # Display the first band, you can change the index if there are multiple bands
    plt.colorbar(label='Pixel Value')  # Add colorbar
    plt.title('Raster Image')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```
